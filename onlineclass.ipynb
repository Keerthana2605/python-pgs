```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(2,8)
y1=x*3
y2=np.array([5,2.6,4,15,20,6])
plt.plot(x,y1,'r')
plt.plot(x,y2,'g')
plt.plot("Multiple line plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x= [7,14,21,28,35,42,49]
y=[5,12,19,21,31,27,35]
z=[3,5,11,20,15,29,31]
plt.plot(x,y,'g',label='Line Y')
plt.plot(x,z,'r', label='Line Z')
plt.legend()
plt.show()
```

```python
# Plotting a Heatmap
#1 . matplotlib imshow()
x=np.arange(100).reshape((10,10))
plt.imshow(x,cmap='rainbow', interpolation='bilinear')
plt.title("Heat map example")
plt.show()
```

```python
#2. matplotlib colormesh function
x=np.arange(10000).reshape((100,100))
plt.pcolormesh(x,cmap='coolwarm')
plt.title("Heat Map example 2")
plt.show()
```

```python
# seaborn library heatmap function - 3rd way
import seaborn as sns
x=np.arange(15**2).reshape((15,15))
sns.heatmap(x,linewidths=0.5, cmap='Dark2')
plt.title("Heat Map ex 3")
plt.show()
```

```python
# plot numpy array as an image
x =np.array([[[0,0,128],[255,255,0], [128,0,0]], [[0,255,0],[0,0,255], [255,0,0]]])
plt.imshow(x)
plt.show()
```

```python
#save plots as images
x=np.array([1,2.5,8,4.3,6])
plt.plot(x)
plt.savefig("numpy_array.png")
plt.show()
```
