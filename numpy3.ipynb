```python
import numpy as np
a = np.array([np.nan,1,2,np.nan,3,4,5])
print(a[~np.isnan(a)])
```

```python
a = np.array([1,2+6j,5,6+7j])
print(a[np.iscomplex(a)])
```

```python
#save and extract numpy
a=np.array([1,2,3,4,5])
np.save('outfile',a)
```

```python
b=np.load('outfile.npy')
print(b)
```

```python
#save it as a text file
a=np.array([1,2,3,4,5])
np.savetxt('out.txt',a)
b=np.loadtxt('out.txt')
print(b)
```

Plotting 

```python
from matplotlib import pyplot as plt
x=np.arange(1,11)
y=2*x+5
plt.title("Matplotlib trial")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=np.arange(1,11)
y=2*x+5
plt.title("Matplotlib trial")
plt.xlabel("X axis")
plt.ylabel("Y axis")
#plt.plot(x,y,"or") #o- circle and r-red(color)
#plt.plot(x,y,":")
plt.plot(x,y,"p")
plt.show()
```

```python
#sine wave
x=np.arange(0,3*np.pi,0.1)
#x=np.arange(0,3,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show
```

```python
#sub plots
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(1,3,1)
plt.plot(x,y_sin)
plt.title("sine")
plt.subplot(2,3,2)
plt.plot(x,y_cos)
plt.title("cos")
plt.show()
```

```python
#bar graph
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color="g",align='center')
plt.title("bar graph")
plt.ylabel("y axis")
plt.xlabel("x axis")
```

```python
#histogram 
a=np.array([22,87,5,23,45,6,7,3,56,45,7,5,34])
plt.hist(a,bins=[0,20,40,60,80,100])
plt.title("histogram trail")
plt.show()
```

```python
x=np.arange(2,8)
y=np.array([5,6,8,20,18,30])
plt.plot(x,y,linestyle="--")
plt.title("array of line")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
```
